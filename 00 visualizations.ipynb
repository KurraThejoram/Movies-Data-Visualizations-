```python
sns.set_style("darkgrid",{"axes.facecolor":"black"})
f,axes = plt.subplots(3,2,figsize=(15,15))
#plot[0,0]
k1=sns.kdeplot(movies['Profitability'],movies['Audience score %'],ax=axes[0,0],cmap='gist_earth',shade=True)
k1=sns.kdeplot(movies['Profitability'],movies['Audience score %'],ax=axes[0,0],cmap='gist_gray_r')

#plot[0,1]
k2=sns.kdeplot(movies['Profitability'],movies['Rotten Tomatoes %'],ax=axes[0,1],cmap='inferno',shade=True)
k2=sns.kdeplot(movies['Profitability'],movies['Rotten Tomatoes %'],ax=axes[0,1],cmap='gist_gray_r')

k1.set(xlim=(-3,13))
k2.set(xlim=(-3,13))
#plot[1,0]
z=sns.violinplot(data=movies,x='Genre',y='Audience score %',ax=axes[1,0],palette='YlOrRd')
#plot[1,1]
k3=sns.kdeplot(movies['Audience score %'],movies['Rotten Tomatoes %'],shade=True, cmap='Blues_r',ax=axes[1,1])
k3=sns.kdeplot(movies['Audience score %'],movies['Rotten Tomatoes %'],cmap='gist_gray_r',ax=axes[1,1])

#plot[2,0]
m1= sns.distplot(movies['Audience score %'],ax=axes[2,0])
#plot[2,1]
axes[2,1].hist(movies['Audience score %'])
```

```python
list1=list()
mylabels=list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre==gen].Profitability)
    mylabels.append(gen)
    
sns.set_style("whitegrid")
fig,ax=plt.subplots()
fig.set_size_inches(15,9)
h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels)
plt.title("Movie Profits",fontsize=35, \
         color='DarkBlue',fontname='Console')
plt.ylabel("No of movies",fontsize=25,color='Red')
plt.xlabel("profits",fontsize=25,color='Red')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(prop={'size':20})
plt.show()
```
